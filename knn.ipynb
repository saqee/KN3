```python
import pandas as pd 

import numpy as np 

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Coded_data.csv')
```

```python
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
df.info()
```

```python
target=df.Result
target
```

```python
feature=df.drop('Result',axis=1)
feature
```

```python
from sklearn.preprocessing import StandardScaler
# Creating instance 'scaler'
scaler = StandardScaler()
```

```python
scaler.fit(feature)
```

```python
scaled_features=scaler.transform(feature)
scaled_features
```

```python
type(scaled_features)
```

```python
df
```

```python
cols=df.columns[:-1]
cols
```

```python
df_scaled_features=pd.DataFrame(scaled_features, columns=cols)
df_scaled_features
```

```python
from sklearn.model_selection import train_test_split

X = df_scaled_features
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) 
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
predictions
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python

print(classification_report(y_test,predictions))
```
